```python
import os
import tempfile
import gradio as gr
from google import genai
from gtts import gTTS

# -----------------------------
# Gemini Client
# -----------------------------
client = genai.Client(
    api_key=os.getenv("GOOGLE_API_KEY")
)

# -----------------------------
# Extract Text
# -----------------------------
def extract_text(file):
    if file is None:
        return "Please upload a file first.", None

    uploaded_file = client.files.upload(file=file.name)

    response = client.models.generate_content(
        model="gemini-2.5-flash",
        contents=[
            uploaded_file,
            "Extract all text exactly as written."
        ]
    )

    return response.text, None


# -----------------------------
# Summarize Document
# -----------------------------
def summarize(file):
    if file is None:
        return "Please upload a file first.", None

    uploaded_file = client.files.upload(file=file.name)

    response = client.models.generate_content(
        model="gemini-2.5-flash",
        contents=[
            uploaded_file,
            "Summarize this document in simple language."
        ]
    )

    return response.text, None


# -----------------------------
# Answer Questions
# -----------------------------
def answer_question(file, question):
    if file is None:
        return "Please upload a file first.", None

    if not question.strip():
        return "Please enter a question.", None

    uploaded_file = client.files.upload(file=file.name)

    response = client.models.generate_content(
        model="gemini-2.5-flash",
        contents=[
            uploaded_file,
            f"Answer the following question using ONLY the uploaded document.\n\nQuestion:\n{question}"
        ]
    )

    return response.text, None


# -----------------------------
# Audio Summary
# -----------------------------
def audio_summary(file):
    if file is None:
        return "Please upload a file first.", None

    uploaded_file = client.files.upload(file=file.name)

    response = client.models.generate_content(
        model="gemini-2.5-flash",
        contents=[
            uploaded_file,
            "Summarize this document in simple language."
        ]
    )

    summary = response.text

    audio_file = tempfile.NamedTemporaryFile(
        suffix=".mp3",
        delete=False
    ).name

    gTTS(text=summary, lang="en").save(audio_file)

    return summary, audio_file


# -----------------------------
# Gradio Interface
# -----------------------------
with gr.Blocks(title="Gemini Document Analyzer") as demo:

    gr.Markdown("# 📄 Gemini Document Analyzer")
    gr.Markdown("Upload a PDF, DOCX, or TXT file and perform AI-powered analysis.")

    file = gr.File(label="Upload Document")

    with gr.Row():
        btn_extract = gr.Button("📄 Extract Text")
        btn_summary = gr.Button("📝 Summarize")
        btn_audio = gr.Button("🔊 Audio Summary")

    gr.Markdown("## ❓ Ask Questions")

    question = gr.Textbox(
        label="Question",
        placeholder="Ask anything about the uploaded document..."
    )

    btn_qa = gr.Button("💬 Answer Question")

    gr.Markdown("---")

    output = gr.Textbox(
        label="Output",
        lines=15
    )

    audio = gr.Audio(
        label="Audio Summary",
        type="filepath"
    )

    btn_extract.click(
        fn=extract_text,
        inputs=file,
        outputs=[output, audio]
    )

    btn_summary.click(
        fn=summarize,
        inputs=file,
        outputs=[output, audio]
    )

    btn_audio.click(
        fn=audio_summary,
        inputs=file,
        outputs=[output, audio]
    )

    btn_qa.click(
        fn=answer_question,
        inputs=[file, question],
        outputs=[output, audio]
    )


# -----------------------------
# Launch App
# -----------------------------
if __name__ == "__main__":
    demo.launch(
        server_name="0.0.0.0",
        server_port=int(os.environ.get("PORT", 10000))
    )
```
